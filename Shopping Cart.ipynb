```python
import os
from IPython.display import clear_output


#------------------------------------------------------------------------------------------------------------------------------#
#--------------------------------------------------------FUNCTIONS-------------------------------------------------------------#            
#------------------------------------------------------------------------------------------------------------------------------#


# create a function that will show the instructions to the program when called upon.
def show_instructions():
    print("Enter 'ADD' to add an item to the list")
    print("Enter 'SHOW' to see your list of items")
    print("Enter 'HELP' to see instructions")
    print("Enter 'CLEAR' to see your list")
    print("Enter 'REMOVE' to delete an item from your list")
    print("Enter 'DONE' when you are finished adding to your list")
    
# create a function that will clear the previous information when called upon. this will keep the page clean.
#def clear_screen():
#    os.system("cls" if os.name == 'nt' else 'clear')

# create a function that will display the current list when called upon.
def display_value():
    for key , value in myDict.items():
        
        # if the user has only one instance of an item in the cart, display only the item name
        if value == 1:
            print(f"{key}")
        
        # if the user amount goes to zero, remove the key entirely from the dictionary
        elif value == 0:
            del key
        
        # if the user has any value that is 2 or greater, display both the item as well as the amount    
        else:
            print(f"{key}: {value}")

            
#------------------------------------------------------------------------------------------------------------------------------#
#----------------------------------------------REST--OF--CODE------------------------------------------------------------------#
#------------------------------------------------------------------------------------------------------------------------------#


# create an empty dictionary to store future values            
myDict = {}

# start out by showing the user what the instructions are
show_instructions()

# setting a while loop to true will cause an infinite loop and allow us to ask questions continually.
while True:
    
    # ask the user what they want to do. this will repeat after every action the user does.
    query = input("What would you like to do? ")
    
    # if the user wants to add we ask them what item they would like to add.
    if query == 'ADD'.lower():
        clear_output()
        query = input("What would you like to add?: ")
        
        # if the item the user wants to submit already is in the dictionary we want to ask if they want to add it again.
        if query in myDict:
            double_check = input("This item already exists. Would you like to add an additional item? Y/N: ")
            
            # if the user answers yes, we add 1 to the current value of the item
            if double_check == "Y".lower():
                myDict[query] = myDict[query] + 1               
            else:
                continue
        
        # if the item the user wants to submit does not already exist in the dictionary we set the value to one.
        else:
            myDict[query] = 1
        
        # we will display the user's list.
        clear_output()
        print("Here is your current list:")
        display_value()
    
    ## if the user wants to display their current list we display the current list.
    if query == 'SHOW'.lower():
        clear_output()
        if myDict: 
            print("Here is your current list:")
            display_value()
        
        # if there is nothing in their list, we let them know that it is empty.
        else:
            print("Whoops, there doesn't seem to be anything in your list!")
    
    # if the user asks to show the instructions, run the function to show the instructions.
    if query == 'HELP'.lower():
        clear_output()
        show_instructions()
    
    # if the user selects clear, ask them if they are sure they want to clear out their entire list.
    if query == 'CLEAR'.lower():
        clear_output()
        areYouSure = input("Are you sure? Press Y/N: ")
        
        # if the user selects yes, we will clear the dictionary out, leaving us with an empty dictionary.
        if areYouSure == 'Y'.lower():
            myDict.clear()
            clear_output()
            print("Your list is now empty")
    
    # if the user selects remove, ask them which item they want to remove.
    if query == 'REMOVE'.lower():
        clear_output()
        whichToRemove = input("Which item would you like to remove?: ")
        
        # if the item they choose is in the dictionary, make sure they want to remove it.
        if whichToRemove in myDict:
            removalCheck = input("Are you sure? Press Y/N: ")
            
            # if they answer yes, take the value(count of item) and subtract it by 1
            if removalCheck == 'Y'.lower():
                myDict[whichToRemove]= myDict[whichToRemove] - 1
                
                # if the value (count of item) goes down to 0, remove the key(item) from the dictionary.
                if myDict[whichToRemove]==0:
                    del myDict[whichToRemove]
                clear_output()
        
        # if the item they enter does not exist in the dictionary, let the user know and display their current items
        else:
            clear_output()
            print("That item doesn't appear to be in the list.\n\nHere is your current list:")
        if myDict:
            display_value()
        
        # if the dictionary is empty, let the user know that they have no items in their list.
        else:
            print("Whoops, there doesn't seem to be anything in your list!")
    
    # if the user selects done, display their current final list.
    if query == 'DONE'.lower():
        clear_output()
        if myDict:
            (print("Here is your list:"))
            display_value()
        
        # if the user has no items in their list, let them know that the list is empty.
        else:
            print("Whoops, there doesn't seem to be anything in your list!")
        break
```
